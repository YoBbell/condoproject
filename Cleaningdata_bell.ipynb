```python
# Extract Number
df['post-by'] = df['post-by'].str.replace("ลงประกาศโดย","").str.strip()
df['constructed-yr'] = df['constructed-yr'].str.extract(r"(\d+)")
df['new'] = df['new'].str.extract(r"(\d+)")
df['price'] = pd.to_numeric(df['price'].str.replace(",",""))
df['space'] = df['space'].str.extract(r'(\d+)')
df['space'] = pd.to_numeric(df['space'])


# Replace same objective column 
# Column constructed-yr and new have similar value, column new have data only 2023 yr
df['constructed-yr'] = df['constructed-yr'].fillna(df['new'])

# Categorize room column
    # Extract wanted part
df['room'] = df['room'].str.findall(r'title="(.+?)">')
    # Categorize function
def split_data(row):
    if isinstance(row, list):
        if len(row) == 2:
            return row[0].split()[0], row[1].split()[0], None
        else:
            if 'สตูดิโอ' in row[0]:
                return None, None, row[0].split()[0]
            else:
                return row[0].split()[0], None, None
    else:
        return None, None, None

    # Apply function
df[['bedroom', 'bathroom', 'studio']] = df['room'].apply(lambda x: pd.Series(split_data(x)))

df['range-from-bts'] = df['range-from-bts'].replace('[]',np.nan)
df['range-from-bts'] = df['range-from-bts'].str.extract(r'(\d+ นาที \(\d+ ม\.\) ถึง .+)')

df[[0,1,2,3,4,5]] = df.name.str.split(',',expand=True)
not_null_rows = df[[5]].notnull().any(axis=1)
df.dropna(subset=[0,1,2,3,4],inplace=True)
df.drop(index=(df[not_null_rows]).index,inplace=True)
df.drop([5],axis=1,inplace=True)

df
```

##### แปลงข้อมูลแขวงเป็น dummy variable

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
dmy = le.fit_transform(cleandata[['kwaeng']])
dmy
```

```python
dk = pd.DataFrame(dmy, columns=["kwaengdmy"])
dk
```

```python
df2 = pd.concat([cleandata, dk], axis='columns')
df2.head()
```

dummy variable
- แขวงบางกะปิ = 0
- แขวงสามเสนนอก = 1
- แขวงห้วยขวาง = 2

```python
print("correlation: ", df2["kwaengdmy"].corr(df2["priceper"]).round(4))
```

ทฤษฎี: correlation (ค่า r) จะมีค่าตั้งแต่ -1 ถึง 1
- ค่า r เข้าใกล้ -1 แสดงว่าตัวแปร 2 ตัวนั้น มีความสัมพันธ์ในเชิงลบ; ถ้า x เพิ่ม y ลด, ถ้า x ลด y เพิ่ม
- ค่า r เข้าใกล้ 1 แสดงว่าตัวแปร 2 ตัวนั้น มีความสัมพันธ์ในเชิงบวก; ถ้า x เพิ่ม y เพิ่ม, ถ้า x ลด y ลด
- ค่า r = 0 แสดงว่าตัวแปร 2 ตัวนั้น ไม่มีความสัมพันธ์กัน

แปลผล: ค่า correlation ระหว่างข้อมูลแขวงที่แปลงเป็น dummy variable แล้ว ได้เท่ากับ 0.07 ซึ่งถึงแม้จะมีค่าเป็นบวก แต่มีค่าที่เข้าใกล้ 0 มาก แสดงว่าตัวแปร "แขวง" และ "ราคาต่อตร.ม." มีความสัมพันธ์กันแบบ weak ในเชิงบวก คือหมายถึง ถ้าเป็นแขวงห้วยขวาง(มีค่า=2) ราคาอาจจะเพิ่มเล็กน้อย, ถ้าเป็นแขวงบางกะปิ(มีค่า=0) ราคาอาจจะลดลงเล็กน้อย

```python
fig = px.box(df2, x='kwaeng', y="priceper")

fig.update_layout(
    title="ที่อยู่ตามแขวง vs ราคาต่อตร.ม.",
    xaxis_title="แขวง",
    yaxis_title="ราคาต่อตร.ม.",
    height=800
)

fig.show()
```

แปลผล: จากกราฟแขวงสามเสนนอกมีราคาที่ต่ำกว่าทั้งห้วยขวางและบางกะปิ สังเกตได้จากค่า mean และ median ที่ต่ำกว่า แต่แขวงห้วยขวางและบางกะปิมีราคาต่อตร.ม.เฉลี่ยไม่ต่างกันมาก และแขวงห้วยขวางมีคอนโดที่มีราคาต่อตร.ม.มากที่สุดที่ราคา 219.6k และมีค่าสุดโต่งที่มากที่สุดเช่นเดียวกันในราคา 279.7k ส่วนแขวงบางกะปิมีคอนโดที่มีราคาต่อตร.ม.ต่ำที่สุดในราคา 13.5k แต่ค่าสุดโต่งที่ต่ำที่สุดอยู่ที่แขวงห้วยขวางที่ราคา 823 บาท
